#### PROYECCIÓN VECTORIAL
La proyección de un vector es la sombra que se proyecta un vector sobre algún eje o sobre otro vector. Para realizar su cálculo supongamos que tenemos dos vectores u y v ambos vectores son diferentes de 0. Entonces la proyección (proy) se describe con la fórmula.


#### DEFINIR UNA UFUNC PARA CALCULAR PROYECCION VECTORIAL SOBRE VECTORES DE NUMPY

La proyeccion se calcula de la forma:






Y su funcion respectiva es la siguiente:


```python
import numpy as np
#funcion de proyeccion

def proyeccion_vectorial(v, u):
    return (np.dot(u,v)/np.dot(v,v)) * v
    
v = np.array([3,1])
u = np.array([1,5])

print(proyeccion_vectorial(v,u))
    
```

Tal como lo explicamos en el markdown anterior, su interpretacion es que la funcion de la proyeccion vectorial nos devuelve el vector resultante que le hace sobre de V a U.
